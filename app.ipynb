```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# User query
user_query = "Create a UI for a gaming app with a modern theme."

# Generate structured prompt
inputs = tokenizer(user_query, return_tensors="pt")
outputs = model.generate(inputs['input_ids'], max_length=50)
prompt = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("Generated Prompt:", prompt)
```

```python
from diffusers import StableDiffusionPipeline

# Load Stable Diffusion
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4").to("cuda")

# Generate image from prompt
image = pipe(prompt).images[0]
image.save("generated_ui.png")
print("UI design saved as 'generated_ui.png'")
```

```python
def generate_ui_design(user_query):
    # Step 1: Generate text prompt
    inputs = tokenizer(user_query, return_tensors="pt")
    outputs = model.generate(inputs['input_ids'], max_length=60)
    prompt = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Step 2: Generate image
    image = pipe(prompt).images[0]
    image.save("ui_design.png")
    return "ui_design.png"

# Main program
user_query = input("Enter your mobile UI design request: ")
output_path = generate_ui_design(user_query)
print(f"Generated UI Design saved at: {output_path}")
```

```python
import gradio as gr

def generate_ui_gradio(user_query):
    inputs = tokenizer(user_query, return_tensors="pt")
    outputs = model.generate(inputs['input_ids'], max_length=50)
    prompt = tokenizer.decode(outputs[0], skip_special_tokens=True)

    image = pipe(prompt).images[0]
    return image

gr.Interface(fn=generate_ui_gradio, inputs="text", outputs="image").launch()
```

```python
from rouge_score import rouge_scorer

def calculate_rouge_score(reference_prompt, generated_prompt):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference_prompt, generated_prompt)
    return scores

# Example Usage
reference_prompt = "Design a user-friendly interface for a mobile banking application."
generated_prompt = "Design a sleek and intuitive UI for a banking app on smartphones."
rouge_scores = calculate_rouge_score(reference_prompt, generated_prompt)
print("ROUGE Scores for Prompt Accuracy:", rouge_scores)
```
